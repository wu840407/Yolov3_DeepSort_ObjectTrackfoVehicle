```python
import os
import cv2

filename=r'./labels/'
def Photo(filename):
    f=open(r'./car.txt',"w+")
    for file in os.listdir(filename):
        fullpath = os.path.join(filename, file)
        if os.path.isdir(fullpath):
            for file1 in os.listdir(fullpath):
                print(file1.split(".")[0])
                f.write("./car/images/"+file+"/"+file1.split(".")[0]+".png\n")
        
    f.close()
if __name__ == '__main__':
    Photo(filename)
```
